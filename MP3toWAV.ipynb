```python
#mp3 into wav 
import glob
                                                                       
lst = glob.glob("*.mp3") # MP3 files  

#convert MP3 into WAV
for file in lst:
	os.system(f"""ffmpeg -i {file} -acodec pcm_u8 -ar 22050 {file[:-4]}.wav""") 
```
