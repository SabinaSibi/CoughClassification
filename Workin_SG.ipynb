```python
# calculating spectrogram 
# spectrogram ploting

import glob

path = 'virufy_renamed'

podaci = glob.glob('.wav')
for korona in podaci:
    fajl = (f"ffwav -i {korona}")
    uzorak, kasalj = wavfile.read(fajl)

    # Plot the spectrogram
    plt.subplot(212)
    plt.title('Spectrogram of  ' + fajl + '  file corona tested')
    plt.specgram(kasalj,Fs=uzorak)
    plt.xlabel('Time [s]')
    plt.ylabel('Frequency [Hz]')
    
plt.show()

#(for Colab save spectrograms into folder on the disk. virufy_spectrograms)
```
