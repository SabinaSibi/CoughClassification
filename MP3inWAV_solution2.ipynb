```python
#convert MP3 into WAV solNo2

import pandas as pd 
import numpy as np 
import os 
import matplotlib.pyplot as plt
import librosa # library for audio processing
import librosa.display 
from glob import glob


!pip install pydub
!apt-get install ffmpeg


from pydub import AudioSegment


mpf='mp3_files'#source MP3 audio file

vaf='wav_files'#destination WAV audio file
                                                    
for file in os.listdir(mpf):
      if file.endswith('.mp3'):
        mp3_file = os.path.join(mpf, file)
        wav_file = os.path.join(vaf, file.replace('.mp3', '.wav'))
        cought = AudioSegment.from_mp3(mp3_file)
        cought.export(wav_file, format='wav')
```
